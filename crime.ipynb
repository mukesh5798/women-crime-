<h1>Women Crime Data 2001 to 2021</h1>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('CrimesOnWomenData.csv')
```

```python
data['State'] = data['State'].str.upper()
```

**7. Find Out Number of State**

```python
data.groupby(['State'])['Rape'].sum()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=data['Rape'], y=data['State'], hue=data['State'], palette='viridis', alpha=0.8)
plt.xlabel('Total Number of Rape Cases', fontsize=13)
plt.ylabel('State', fontsize=13)
plt.xticks(rotation=60)
plt.title('Number of Rape Cases by State', fontsize=15)
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
